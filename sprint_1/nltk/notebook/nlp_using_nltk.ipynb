**3.Calculate a measure of the lexical richness of the text**
* Lexical richness is about the quality of vocabulary in a language sample.
* Lexical richness = len(set(text))/len(text)

```python
# lexical richness for text3
num_distinct_words = len(set(text3))
num_total_words = len(text3)
                         
lexical_richness = num_distinct_words/num_total_words
print(f"Lexical richness of **text3** is {lexical_richness}")
```

* Higher the number of distinct words in given text, higher the lexical richness will be.
* From above lexical richness is 0.062, i.e number of distinct words is just 6% of the total number of words.

Now let's define fuction named lexical_diversity() to calculate lexical richness of text


```python
def lexical_diversity(text):
    '''
    calculates lexical richness of text
    
    Arguments:
    text: text whose lexical richness to be calculated
    
    Returns:
    lexical_richness: calculates lexical_richness of text
    '''
    lexical_richness = len(set(text))/len(text)
    return lexical_richness
```

* Frequency Distribution tells us the frequency of each vocabulary item in the text.
* It is a "distribution" because it tells us how the total number of word tokens in the text are distributed across the vocabulary items.
* Frequency Distribution is often needed in language processing, NLTK provides built-in support for them
* Let's see FreqDist,



```python
fdist = FreqDist(text1)
```

**Collocations and Bigrams**
* A collocation is a sequence of words that occur together unsually often, Thus red wine is a collocation, whereas the wine is not.
* Also, collocations are phrases or expressions that are highly likely to co-occur
* We extract frequently together words from the text
* Example: good film, las vegas, new york and so-on
* Visit for more:
https://towardsdatascience.com/collocations-in-nlp-using-nltk-library-2541002998db
https://www.nltk.org/howto/collocations.html
**How collocations different from Bigrams**

* Bigrams means the set of two words(e.g. this is, red wine, is said and so-on) that co-occur, while trigrams means set of three words that co-occur
* Bigrams or Trigrams may not give us meaning phrases.
* Example:
*  consider sentence: "He applied machine learning" contains biagrams i.e.
*  Biagrams: 'He applied', 'applied machine', 'machine learning'.
*  'He applied' and 'applied machine' is not meaningful, while 'machine learning' is a meaningful biagrams.
* Thus, just considering co-occurring words may not be a good idea, since phrases such as 'of the' may occur frequently, but are actually not meaningful
* Thus the need for collocations from NLTK library. It only gives us the meaningful bigrams and trigrams
**How one collocation better than other?**

* Pointwise Mutual Information or PMI score is used.
* PMI scores is used to rank the bigrams and trigrams churned out by collocations library.
*  PMI(a, b) = log( p(a, b)/(p(a) * p(b)) )

# Counting Other Things Than Words

* Counting words is useful, but we can count other things too.
* Example: We can look at the distribution of word lengths in text, shown below,

```python
#Distribution of word lengths in text1  

#make list of length of all words
word_length_list = [len(word) for word in text1]

#generate frequency distribution using FreqDist
fdist = FreqDist(word_length_list)
print(fdist)

#show frequency  distribution
fdist
```

* As we can see from above, text1 contains maximum words with length=3
* Whereas, text1 contains minimum words with length=3

```python
fdist.plot(20, cumulative = False)
plt.show()
```

```python
fdist.plot(20, cumulative = True)
plt.show()
```

```python
#find most common
fdist.most_common()
```

```python
#find word length whose count is maximum
fdist.max()
```

```python
#%word of length 3
print("The % of word in text1 of length 3 is:", fdist.freq(3))
```

**Word comparison operators**
* Helps to select words from our texts.
* Some word comparisons operators are listed below
* s.startswith(t):	test if s starts with t
* s.endswith(t):	test if s ends with t
* t in s:	test if t is a substring of s
* s.islower():	test if s contains cased characters and all are lowercase
* s.isupper():	test if s contains cased characters and all are uppercase
* s.isalpha():	test if s is non-empty and all characters in s are alphabetic
* s.isalnum():	test if s is non-empty and all characters in s are alphanumeric
* s.isdigit():	test if s is non-empty and all characters in s are digits
* s.istitle():	test if s contains cased characters and is titlecased (i.e. all words in s have initial capitals)
* Demonstration using above is as shown,



```python
#list of words ending with -ableness- from Text1
print(sorted(w for w in set(text1) if w.endswith('ableness')))
```

```python
#list of words containing -gnt-
print(sorted(term for term in set(text4) if 'gnt' in term))
```

```python
#list of words having an initial capital
print(sorted(item for item in set(text6) if item.istitle()))
```
